```python
import pandas
chicago_socioeconomic_data = pandas.read_csv('https://data.cityofchicago.org/resource/jcxq-k9xf.csv')
chicago_socioeconomic_data
```

```python
# Create SQLAlchemy engine
engine = create_engine('postgresql://postgres:1234@localhost:5432/INSTRUCTOR')

# Convert DataFrame to SQL table
chicago_socioeconomic_data.to_sql('chicago_socioeconomic_data', engine, if_exists='replace', index=False)
```

```python
chicago_socioeconomic_data = %sql SELECT * FROM test_score_distribution
chicago_socioeconomic_data
```

```python
%sql SELECT count(*) FROM chicago_socioeconomic_data
```

```python
%%sql
SELECT COUNT(*), community_area_name, hardship_index
FROM chicago_socioeconomic_data
WHERE LOWER(community_area_name) LIKE '%chicago%' AND hardship_index > 50
GROUP BY community_area_name, hardship_index;

```

```python
%sql SELECT count(*) FROM chicago_socioeconomic_data WHERE hardship_index > 50.0
```

```python
%sql SELECT MAX(hardship_index) FROM chicago_socioeconomic_data
```

```python
%%sql
SELECT community_area_name,hardship_index  FROM chicago_socioeconomic_data WHERE hardship_index = (SELECT MAX(hardship_index) FROM chicago_socioeconomic_data)
```

```python
%%sql
SELECT community_area_name, per_capita_income_ FROM chicago_socioeconomic_data WHERE per_capita_income_>60000
```

```python
%%sql
SELECT * FROM chicago_socioeconomic_data ORDER BY per_capita_income_ DESC NULLS LAST LIMIT 10 
```

```python
%sql SELECT * FROM chicago_socioeconomic_data ORDER BY per_capita_income_ ASC FETCH FIRST 5 ROWS ONLY
```

```python
%%sql 
SELECT REPLACE(community_area_name,' ','_') as "community_area_name_without_space" FROM chicago_socioeconomic_data 
ORDER BY percent_of_housing_crowded
FETCH FIRST 5 ROWS ONLY
```

```python
query = %sql SELECT per_capita_income_, hardship_index FROM chicago_socioeconomic_data
df = query.DataFrame()
%matplotlib inline
# uncomment the following line if you get an module error saying seaborn not found
# !pip install seaborn
import seaborn

plot = seaborn.scatterplot(x='per_capita_income_',y='hardship_index', data=df)
```
